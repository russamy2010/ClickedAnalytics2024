The `game_df` table contains game-level data, including the home and away teams, various game statistics, and the season type. We can use this table to create a CTE that calculates the average points scored by the home team for each season and season type combination.

````
from pandas import DataFrame

# Filter the game_df to only include rows where season_type is 'Regular Season'
regular_season_df = game_df[game_df['season_type'] == 'Regular Season'].copy()

# Group the filtered DataFrame by `season_id` and `season_type` and calculate the mean of `pts_home`
avg_home_points_by_season_df: DataFrame = regular_season_df.groupby(['season_id', 'season_type'])['pts_home'].mean().reset_index()

# Rename `pts_home` column to `avg_pts_home`
avg_home_points_by_season_df = avg_home_points_by_season_df.rename(columns={'pts_home': 'avg_pts_home'})

# Create a CTE called `avg_home_points_by_season` using this new dataframe
with pd.option_context("display.max_rows", None, "display.max_columns", None):
  avg_home_points_by_season = avg_home_points_by_season_df.assign(
      avg_pts_home=lambda x: x["avg_pts_home"].round(2)
  )
  cte_query = f"""
  WITH avg_home_points_by_season AS (
    {avg_home_points_by_season.to_markdown(index=False, numalign="left", stralign="left")}
  )
  SELECT *
  FROM avg_home_points_by_season
  """

  # Display the CTE
  print(cte_query)
````
